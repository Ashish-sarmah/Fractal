```python
from math import sqrt
nsr=[]
for k in range(len(img1)):
  lii1=[]
  a,b=img1[k].shape
  l=int((a**2+b**2)**0.5)
  for i in range(1,l+1):
    lii1.append(i)
  nsr.append(lii1)

print(nsr)
```

```python
npn=[]
nmsid=[]
for k in range(len(img1)):
  npn1=[]
  nmsid1=[]
  for i in range(len(nsr[k])):
    npn1.append(0)
    nmsid1.append(0)
  npn.append(npn1)
  nmsid.append(nmsid1)
print(len(npn[0]),len(nmsid[0]))
print(npn)
print(li[0])
```
